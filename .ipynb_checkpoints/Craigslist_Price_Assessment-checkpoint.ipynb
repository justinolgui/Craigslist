```python
import requests
import pandas as pd
import bs4
```

```python
# Setting our search parameter
search = "phone"
```

```python
# The sorting options are relevance, price ascending and price descending
sort = "rel"
```

```python
# Build our search
init_search = "https://vancouver.craigslist.org/search/sss?sort=" + sort + "&query=" + search
```

```python
# Get all prices for our search on craigslist
r = requests.get(init_search)
exampleSoup = bs4.BeautifulSoup(r.text, 'html.parser')
```

```python
# Loop through search results and return prices
price_elem = exampleSoup.find_all('span', {'class':"result-price"})

for i in price_elem:
    print(i.text)
```

```python
# Loop through search results and return location
title = exampleSoup.find_all('a', {'class':"result-title hdrlnk"})

for i in title:
    print(i.text)
```

```python
# Loop through search results and return location
location = exampleSoup.find_all('span', {'class':"result-hood"})

for i in location:
    print(i.text)
```
